```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import j1
from scipy.special import j0
from scipy.special import jn_zeros
import pandas as pd
import sympy as sp
import mpmath as mp
from scipy.integrate import quad
%matplotlib inline
```

```python
bessel_zeros = jn_zeros(0,300) # 500 zeros of bessel function arranged in assending order 
```

```python
def TemperatureDistrubutionCylinderQ0(r,maxRadius,l,maxLength,terms,T):
    a = maxRadius
    L = maxLength
    bessel_zeros = jn_zeros(0,500)
    sol = 0
    
    def am0(m,bessel_zeros,T,terms1):
        j00m = bessel_zeros[m-1]
        term1 = 0 
        for i in range(terms1):
            term1 += (((-1)**i)*((j00m/2)**(2*i)))/(mp.factorial(i+1)*mp.factorial(i))
        term2 = (mp.sinh(j00m/a)*(j1(j00m))**2)
        return (T*term1)/term2
    
    for i in range(1,terms+1):
        term1 = am0(i,bessel_zeros,T,100)
        j00m = bessel_zeros[i-1]
        term2 = j0((r/a)*j00m)*mp.sinh((L/a)*j00m)
        sol += term1*term2
    
    return sol
        
        
    
```

```python
def TemperatureDistrubutionCylinderQ0(r,maxRadius,z,maxLength,terms,T):
    a = maxRadius
    L = maxLength
    bessel_zeros = jn_zeros(0,500)
    sol = 0
    
    def numeriacalintegrration(a,alpha0m):
        def f(x):
            return j0(x*alpha0m/a)*x
    res, err = quad(f, 0, a,limit = 1000)
    return res
    
    def am0(a,m,bessel_zeros,T):
        alpha0m = bessel_zeros[m-1]
        term1 = numeriacalintegrration(a,alpha0m)
        term2 = (mp.sinh(alpha0m/a)*(j1(alpha0m))**2)
        return (2*T*term1)/(term2*a**2)
    
    for i in range(1,terms+1):
        term1 = am0(a,i,bessel_zeros,T)
        alpha0m = bessel_zeros[i-1]
        term2 = j0((r/a)*alpha0m)*mp.sinh(((L-z)/a)*alpha0m)
        sol += term1*term2
    
    return sol
```
