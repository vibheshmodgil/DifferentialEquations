```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import i0
import pandas as pd
import sympy as sp
%matplotlib inline
```

```python
T0, T1, T2, k, z, L, theta = sp.symbols('T0 T1 T2 k z L theta')
```

```python
n,m = sp.symbols('n m', integer = True)
expr = (T2  + (T1-T2)*(z/L))
expr
```

```python
f = (2/L)*(T0-expr)*sp.sin((m*sp.pi*z)/L)
f
```

```python
a = sp.integrate(f ,(z, 0, L))
```

```python
sp.simplify(a)
```

```python
def tempDistributionCylinderOne(r,maxRadius,z,axialLength,radialSurfaceTemperature,upperZtemp,lowerZtemp,noOfTerms) :
    T0 = radialSurfaceTemperature
    T1 = upperZtemp
    T2 = lowerZtemp
    L = axialLength
    a = maxRadius
    n = noOfTerms
    
    def term1(n,T0,T1,T2):
        return (2*((T0)*(1-(-1)**n) + ((-1)**n)*T1 - T2))/(np.pi*n)
    
    b = np.zeros(n)
    for i in range(1,n):
        b[i] =  (term1(i,T0,T1,T2))*(i0(i*r)/i0(i*a))*np.sin((i*np.pi*z)/L)
        
    return b.sum() + (T1-T2)*(z/L) + T2
   
```

```python
r = 0.273/2
maxRadius = 0.273/2 
z = 0.068/2
axialLength = 0.068
radialSurfaceTemperature = 300
upperZtemp = 310
lowerZtemp = 320
k = []
for i in range(1,5000) :
    k.append(tempDistributionCylinderOne(r,maxRadius,z,axialLength,radialSurfaceTemperature,upperZtemp,lowerZtemp,i))
x = np.arange(1,5000)
plt.plot(x,k)
plt.title("Effect of No. of terms \nPoint Choosen is at mid length")
plt.xlabel("Temperature")
plt.ylabel("No. of Terms")
plt.show()
```

```python
r = 0.273/2 
maxRadius = 0.273/2 
axialLength = 0.068
radialSurfaceTemperature = 300
upperZtemp = 310
lowerZtemp = 320
noOfTerms = 5000
zi = np.linspace(0,axialLength,100)
p = []
for i in zi :
    p.append(tempDistributionCylinderOne(r,maxRadius,i,axialLength,radialSurfaceTemperature,upperZtemp,lowerZtemp,noOfTerms))
plt.plot(zi,p)
plt.title("Temperature of Curved Surface")
plt.xlabel("Axial Length")
plt.ylabel("Temperature")
plt.show()
```
